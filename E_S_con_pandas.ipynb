# Aula 5: Creando y leyendo bando de datos

## Creando un banco de datos

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

## Escribiendo en un banco de datos

```python
import pandas as pd
```

```python
archivo = 'clientes_banco.csv'
```

```python
datos_clientes_banco = pd.read_csv(archivo)
```

```python
datos_clientes_banco.head()
```

```python
datos_clientes_banco.to_sql('clientes', engine, index=False)
```

Me permite revisar lo que existe en la base de datos

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

## Leyendo una consulta SQL

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renta = "Empleado"'
```

```python
empleados = pd.read_sql(sql = text(query),con = engine.connect())
```

```python
type(empleados)
```

```python
empleados.head()
```

```python
empleados.to_sql('empleados', con = engine.connect(), index=False)
```

```python
print(inspector.get_table_names())
```

```python
pd.read_sql_table('empleados', con = engine.connect())
```

```python
pd.read_sql_table('empleados', con = engine.connect(), columns=['ID_Cliente', 'Grado_estudio'])
```

## Actualizando banco de datos

```python
query = 'SELECT * FROM empleados'
```

```python
pd.read_sql(sql = text(query), con = engine.connect())
```

####Borrando registros

```python
from sqlalchemy.exc import SQLAlchemyError
query = "DELETE FROM clientes where ID_Cliente = 5008804"

try:
  r_set = engine.connect().execute(text(query))

except SQLAlchemyError as e:
  print(e)
else:
  print('Registro borrado: ', r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con = engine.connect())
```

####Actualizar registro

```python
query = "UPDATE clientes set Grado_estudio = 'Nivel superior' where ID_Cliente = 6842765"

try:
  r_set = engine.connect().execute(text(query))

except SQLAlchemyError as e:
  print(e)
else:
  print('Registro actualizado: ', r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con = engine.connect())
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```
